```python
#Area between two-lines
import matplotlib.pyplot as plt
import numpy as np
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color='green')
plt.plot(time,expense,color='red')
plt.fill_between(time,income,expense,where=(income>expense),color='green',alpha=0.25,label='positive',interpolate=True)
plt.fill_between(time,income,expense,where=(income<expense),color='red',alpha=0.25,label='positive',interpolate=True)
plt.legend()
plt.show()
```

```python
#PIE-Chart
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10))
y=np.array([35,25,25,15])
mylabels=['Apple','Banana','Cherry','Dates']
myexplode=[0,0.2,0,0.5]
mycolors=['black','blue','yellow','brown']
plt.pie(y,startangle=90,labels=mylabels,explode=myexplode,colors=mycolors)
plt.title("FRUITS")
plt.legend(title='FOUR FRUITS')
plt.show()
```
