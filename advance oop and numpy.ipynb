```python
#OVERLOADING AND OVERRIDDING
def sum_num(*args):
    result=0
    for num in args:
        result+=num
    print("RESULT =",result)
print("SIMILER TO METHOD OVERLOADING",)
print("SINGLE ARGUMENT ->",end=" ")
sum_num(10)
print("TWO ARGUMENT ->",end=" ")
sum_num(10,20)
print("MULTIPLE ARGUMENT ->",end=" ")
sum_num(10,20,30,40,50)
```

```python
#MATHEMATICAL OPERATORS
class A:
    def __init__(self,a):
        self.a=a
    def __ge__(self,other):
        if self.a>=other.a:
            return True
        else:
            return False
    def __le__(self,other):
        if self.a<=other.a:
            return True
        else:
            return False
    def __truediv__(self,other):
        return self.a/other.a
    def __floordiv__(self,other):
        return self.a//other.a

obj1=A(5)
obj2=A(4)
obj3=A(10)
obj4=A(20)
obj5=A(30)
if obj1>=obj2:
    print("obj1 is equal or greater")
else:
    print("obj2 is equal or greater")
if obj1<=obj2:
    print("obj1 is equal or less then")
else:
    print("obj2 is equal or less then")
print(obj3/obj2)
print(obj3//obj2)
print(obj5/obj4)
print(obj5//obj4)
```

```python
# METHOD RESOLUTION ORDER
class A:
    def rk(self):
        print("Class A")
class B(A):
    def rk(self):
        print("Class B")
class C(A):
    def rk(self):
        print("Class C")
class D(B,C):
    pass

r=D()
r.rk()
print(D.__mro__)
print(D.mro())
```

```python
class p1:
    def foo(self):
        print("foo from p1")
class p2:
    def foo(self):
        print("foo from p2")
    def bar(self):
        print("bar from p2")
class c1(p1,p2):
    pass
class c2(p2):
    def bar(self):
        print("bar from c2")
class GC(c1,c2):
    pass

obj=GC()
obj.bar()
obj.foo()
print(GC.__mro__)
```

```python
#Abstract class
from abc import ABC,abstractmethod
class shape(ABC):
    def __init__(self,shape_name):
        self.shape_name=shape_name
    @abstractmethod
    def draw(self):
        print("Drawing :".self.shape_name)

class circle(shape):
    def __init__(self):
        super().__init__("Circle")
    def draw_circle(self):
        print("Drawing :",self.shape_name)
    def draw(self):
        pass
        
obj=circle()
obj.draw_circle()
```

```python
#class book -:name,,no ot authors,list of authors,publisher,isbn,year
#class coursr_book(book) -: +course_name
#overrides display method
class Book:
    def __init__(self,name,noofaut,listofaut,pub,isbn,year):
        self.name=name
        self.noofaut=noofaut
        self.listofaut=listofaut
        self.pub=pub
        self.isbn=isbn
        self.year=year
    def display(self):
        print("BOOK NAME :",self.name)
        print("NO OF AUTHORS :",self.noofaut)
        print("LIST OF AUTHORS :",self.listofaut)
        print("PUBLISHER :",self.pub)
        print("ISBN :",self.isbn)
        print("YEAR :",self.year)
        
class course_book(Book):
    def __init__(self,course_name,name,noofaut,listofaut,pub,isbn,year):
        super().__init__(name,noofaut,listofaut,pub,isbn,year)
        self.course_name=course_name
    def display(self):
        super().display()
        print("COURSE NAME :",self.course_name)
        
l=["ABC","XYZ"]
c=course_book("CSE","PYTHON1",2,["XYZ","ABC"],"PQR",1234,1999)
c.display()
```

```python
class student:
    def __init__(self,rollno,name,age,total_marks):
        self.rollno=rollno
        self.name=name
        self.age=age
        self.total_marks=total_marks
    def display(self):
        print("ROLL NO:",self.rollno)
        print("NAME :",self.name)
        print("AGE :",self.age)
        print("TOTAL MARKS :",self.total_marks)
    def __eq__(self,other):
        if self.total_marks==other.total_marks:
            return True
        else:
            return False
    
s1=student(1,"JAIV",20,98)
s2=student(2,"HARSHIT",21,98)
if s1==s2:
    print("EQUAL")
else:
    print("NOT EQUAL")
```
