```python
f=open('text2.txt','r')
f1=open('text.txt','w')
r=f.readlines()
for i in r:
    if i[0]=='#':
        continue
    else:
        a=i.find('#')
        if a==-1:
            f1.write(i)
        else:
            f1.write(i[:a:]+'\n')
            
f.close()
f1.close()
        

```

```python
#wap to count the frequency of words in given text file.
f=open("demo.txt",'r')
a=f.readlines()
l={}
for i in a:
    b=i.split()
    for j in b:
        if j not in l:
            l[j]=1
        else:
            l[j]+=1
        
print(l)
for m,n in l.items():
    print(m,'-',n)
        
```

```python
#wap to read line by line from a given file1 and file2  write into file 3
f1=open("data.txt",'r')
f2=open("studentdata.txt",'r')
f3=open("new1.txt",'w')

l1=len(f1.readlines())
l2=len(f2.readlines())
f1.seek(0)
f2.seek(0)
l=max(l1,l2)
a1=0
a2=0
for i in range(l):
    if a1!=l1:
        f3.write(f1.readline())
        a1+=1
    if a2!=l2:
        f3.write(f2.readline())
        a2+=1

f1.close()
f2.close()
f3.close()
```

```python
#waP to read a text file and print text file with where words starting from i should be printed in reverse order
f=open('demo.txt','r')
a=f.readlines()
for i in a:
    s=''
    b=i.split()
    for j in b:
        if j[0]=='i' or j[0]=='I':
            s+=j[::-1]+" "
        else:
            s+=j+" "
    print(s)
```
